```python
(trainX, trainY), (testX, testY) = tf.keras.datasets.mnist.load_data()
```

Convert Output label to multiple values

```python
print(trainY.shape)
print('First 2 examples are: ', trainY[0:2])
```

```python
#because the label has 10 different possible classes we change them tf's ohe
trainY = tf.keras.utils.to_categorical(trainY, num_classes=10)
testY = tf.keras.utils.to_categorical(testY, num_classes=10)
```

```python
trainY[0]
```

```python
trainY.shape
```

```python
trainX.shape
```

```python
#we then scale our data. max pixel is 255
trainX = trainX/255 
testX = testX/255
```

```python
trainX[0]
```

```python
pd.DataFrame(trainX[2])
```

Build the Graph

```python
#Initialize Sequential model
model = tf.keras.models.Sequential()

#Reshape data from 2D to 1D -> 28x28 to 784
model.add(tf.keras.layers.Reshape((784,),input_shape=(28,28,)))

model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(50, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='relu'))
#Add Dense Layer which provides 10 Outputs after applying softmax for output layer for multi_class classification
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
sgd_op = tf.keras.optimizers.SGD(learning_rate=0.01)
#Comile the model
model.compile(optimizer=sgd_op, loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

Execute the Graph

```python
model.fit(trainX, trainY, 
          validation_data=(testX, testY), 
          epochs=5)
```

Save Model

```python
model.save('mnist_lc.h5')
```
