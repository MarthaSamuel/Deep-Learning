```python


# Preprocess the data
train_images = train_images.reshape((60000, 28 * 28)).astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28)).astype('float32') / 255
```

```python
# Add bias input to the images
train_images = np.hstack([np.ones((60000,1)), train_images])
test_images = np.hstack([np.ones((10000,1)), test_images])
```

```python
learning_rates = [0.001, 0.01, 0.1]
# Initialize weights randomly between [-0.05, 0.05]


# Initialize dictionaries to store accuracies for each learning rate
train_accuracies = {eta: [ ] for eta in learning_rates}
test_accuracies = {eta: [ ] for eta in learning_rates}


for eta in learning_rates:
    weights = np.random.uniform(-0.05, 0.05, (10, 785))

    # Training loop for each epoch
    for epoch in range(70):  # Number of epochs
        correct_predictions = 0#keeps count for accuracy purposes


        # Loop through all training examples
        for i in range(len(train_labels)):
            xk = train_images[i]
            target_label = train_labels[i]

            # Compute output values for all perceptrons
            outputs = np.dot(weights, xk)

            # Predicted label is where output is maximum
            predicted_label = np.argmax(outputs)

            # Update weights if prediction was incorrect
            if predicted_label != target_label:
                weights[target_label] += eta * xk
                weights[predicted_label] -= eta * xk
            else:
                correct_predictions += 1


        # Calculate accuracy for the epoch
        accuracy = correct_predictions / len(train_labels) * 100
        train_accuracies[eta].append(accuracy)
        print(f"Epoch {epoch}: Accuracy {accuracy}%")

#for test
        correct_predictions = 0
        test_predictions = []  # List to store predictions for confusion matrix

        for i in range(len(test_images)):
            xk = test_images[i]
            predicted_label = np.argmax(np.dot(weights, xk))
            if predicted_label == test_labels[i]:
                correct_predictions += 1
            test_predictions.append(predicted_label)


        test_accuracy = correct_predictions / len(test_images) * 100
        test_accuracies[eta].append(test_accuracy)

        print(f"Test Accuracy: {test_accuracy}%")


    plt.figure(figsize=(10, 6))  # Create a new figure for each learning rate

    # Plot training accuracy
    plt.plot(train_accuracies[eta], label=f'Training LR={eta}')

    # Plot test accuracy
    plt.plot(test_accuracies[eta], label=f'Test LR={eta}', linestyle='--')

    plt.title(f'Accuracy over Epochs (LR={eta})')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    # Compute and plot the confusion matrix for the last epoch
    conf_matrix = confusion_matrix(test_predictions, test_labels)
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix (LR={eta})')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()


```
